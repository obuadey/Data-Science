```python
import numpy as np
import csv
import pandas as pd
import datetime

with open('AAPL-2017.csv', newline = '') as a:
    reader = csv.DictReader(a)
    stock2017 = np.empty((0, 2), dtype=str)
    for row in reader:
        oneRow = np.array([[row['Date'], row['Volume']]])
        stock2017 = np.append(stock2017,oneRow, axis=0)

print(stock2017)
```

```python
date = [datetime.datetime.strptime(str(x),'%Y-%m-%d') for x in stock2017[:,0]] #convert a str date to a date format
```

```python
print(date)
```

```python
data = pd.Series(np.float64(stock2017[:,1]), index=date) # convert the data to a pandas series from numpy, the float is used to specify the type because you cant plot string values
```

```python
print(data)
```

```python
data.plot() #this plots the data 
```
